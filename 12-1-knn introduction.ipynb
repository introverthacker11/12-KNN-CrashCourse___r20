For finding closest similar point, you find the distance between points using distance measures such as:
* Euclidean Distance
    * Euclidean Distance is most popular distance measure.
    * Euclidean Distance is the shortest distance between two points.
* Manhattan Distance
    * Manhattan Distance is the distance between two points in a grid.
    * It measures sum of absolute differences between two points.
* Minkowski Distance
    * Minkowski Distance is the generalized form of Euclidean and Manhattan Distance.
    * It takes the sum of absolute differences to the power of p. where p is a positive integer, which is also called as order of norm.
* hamming Distance
    * Hamming distance is the distance between two points in a binary space.
    * It measures the number of positions at which the corresponding symbols are different.
    * Hamming distance is used in DNA sequencing.
    * Hamming distance is also used in spell checking.
    * Hamming distance is also used in error correction.
    * Hamming distance is also used in data compression.

**What is K-Nearest Neighbors (KNN)?**  
K-Nearest Neighbors (KNN) is a machine learning algorithm that helps us **classify** or **predict** something based on what is nearby (or similar). It's like asking your **neighbors** for advice, and you make a decision based on what most of them say.

**How does KNN work?**

1. **K is the number of neighbors**: In KNN, "K" represents the number of **neighbors** we look at to make a decision.
   - If **K = 1**, the algorithm looks at **one** nearest point to make a prediction.
   - If **K = 3**, it looks at the **three** nearest points and then decides based on what the majority of the points say.

**Example:**
- Imagine you're in a **new neighborhood** and you want to find out what type of **food** people like.
- You ask your **3 closest neighbors** (K = 3). If **2 of them** like **pizza** and 1 likes **sushi**, you predict that the area is more of a **pizza** neighborhood.

**How to make a prediction using KNN:**
- Suppose you have a **point (P1)** and you want to predict its label (like classifying if its a dog or a cat).
   - You check how far other points (neighbors) are from P1.
   - If **most nearby points** are cats, you predict P1 is a cat.
   - If **most are dogs**, you predict P1 is a dog.

**Choosing K:**
- If **K = 1**, you ask just one neighbor (the closest one). 
- If **K = 3**, you ask three neighbors and go with the **majority**.
- If **K = 7**, you ask seven neighbors.

#### Example with K=1, K=3, and K=7:
- In the image (example), if **K=3**, it looks at the **three closest neighbors**.
- If **K=7**, it looks at the **seven closest neighbors**. 

The bigger the circle, the more neighbors you consider.

**Distance Measures:**  
To decide which points are closest, we use **distance**. Here are different ways to measure distance:

1. **Euclidean Distance** (most common):
   - This is the **straight-line distance** between two points, like the shortest path.
   - **Example**: If you walk straight across a park to your friend, thats Euclidean distance.
   
2. **Manhattan Distance**:
   - This is the distance between two points when you can only move in **right angles**, like city blocks.
   - **Example**: Imagine you're walking along streets in a city, turning corners as you gothis is like walking in a grid.

3. **Minkowski Distance**:
   - This is a **general form** of both Euclidean and Manhattan distances.
   - **Example**: It adjusts how we calculate distance by using a power (**p**) to change the rules.

4. **Hamming Distance**:
   - This measures the difference between **two sets of symbols**, often used when comparing sequences.
   - **Example**: If you compare two words, cat and bat, Hamming distance counts how many letters are different (here, one difference: "c" vs. "b").

   - **Where its used**: Its used in fields like DNA sequencing or spell checkers. For example, comparing **DNA** sequences to see where they differ.

**In Summary:**
- **K** is the number of neighbors you look at to make a decision.
- You calculate how **close** or **far** your data points are using distance measures (like **Euclidean** or **Manhattan**).
- The KNN algorithm uses the **closest neighbors** to predict or classify something based on how similar they are.

By asking your neighbors (data points), you can make a decision about new data based on what nearby points have already shown!
